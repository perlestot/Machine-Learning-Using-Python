# Methods
This study consists of 10 processes. The details are described follows;

1. **Data acquisition**<br>
The collection of data in this study is retrieved from social media call "honestdocs.co" with the score range from 1(dissatisfaction) to 5 (satisfaction).

2. **Translate all customers' comments into English**<br>
There are many tools for text processing in English, such as excellent NLTK, simple TextBlob, or the fastest syntactic parser spaCy. Unfortunately, there are a few tools for Thai language. Therefore, we decides to translate all comment into English using Google translate API.  

3. **Deduplicate comments and elimination the row, which contains null value.**<br>

4. **Text preprocessing**<br>
After collecting the data, the preprocessing phase allows to reduce noise in data includes normalizing text to the same Unicode, lowering case, removing punctuation and emoticons, expanding contraction, removal of Stop words, word lemmatizations.

5. **Encoding** <br>
For sentiment analysis, we would like to know what is make customer dissatisfaction and satisfaction. Hence, we will encode comment with score 1s, 2s, and 3s to the negative class and score 4s and 5s to the positive class.

6. **Split data**<br>
For developing sentiment classifier, the author will split data into two sets, which are training set (training data and validation data) and testing set with stratifying class. The new model is developed on the training set and test it on the holdout set.

7. **Feature extraction**<br>
Features play a key role in determining the performance of the model. In this study, we will transform text to sequence and use dense embeddings for feature extraction.

8. **Developing model and Fine-tuning hyperparameter**<br>
In this study, we will develop a model from recurrent neural network (RNN) architecture (LSTM; Long short-term memory). The model hyperparameters are tuned by using randomized-search and grid-search based on accuracy with using <u>class weight</u> to get the best parameter. 

9. **Evaluation**<br>
After the model is built, we will evaluates it with testing set by using <u>balanced accuracy</u>  and other metrics such as precision on negative class, accuracy, and F1 to estimate the performance. The model with the highest performance is chosen.

10. **Syntactic analysis and visualization** <br>
To find what is dissatisfaction and satisfaction of the patient, we have to know what is the topic customer said. The final step is syntactic of each comment, which analyzed using POS tagging and shallow parsing. All noun phrase is extracted and presented as the comment topic. For visualizing comment topic separated by the sentiment prediction (i.e., positive and negative). All topic will be summarized and presents in bag of word form (term-frequency). This information will be visualized by word cloud, and the topic sizes represent term frequency (i.e., the bigger font means the more frequency topic).

### Class Weight
<p style="text-indent: 2.5em;">
    Intuitively, we want to give higher weight to minority class and lower weight to majority class. scikit-learn has a convenient utility function to calculate the weights based on class frequencies.

### Balanced Accuracy
<p style="text-indent: 2.5em;">
    The balanced accuracy in binary and multiclass classification problems to deal with imbalanced datasets. It is defined as the average of recall obtained on each class. On the other word, it is avearge of combining sensitivity and specificity.

### Handling overfitting
<p style="text-indent: 2.5em;">
    <b>Overfitting</b> means the data is fit to training data which can make high performance but weak for testing data. Especially on LSTM and small data can be easily overfitting to the data by memorize feature and predict the class. Therefore, we have to handle this issue, and we can do it with three options below;

1. Reduce the network’s complexity by removing layers or reducing the number of elements in the hidden layers
- Apply regularization, which add a component that will penalize large weights.
- Using dropout layers or increase drop rate, which will randomly remove certain features by setting them to zero

### Develop Model with Best Parameter

```python
def build_model(total_comment = df_all['preprocessed'].values, n_hidden_layer=1, 
                n_neurons=16, optimizer_mode='Adam', dropout_rate=0.5, recurrent_dropout_rate=0.5): 
    
    # Find the number of unique words in the data
    total_word = []
    for sent in total_comment:
        total_word.extend(sent.split())
    max_feature = len(set(total_word))
    max_length = max([len(sent.split()) for sent in total_comment]) # Equal to X.shape[0]

    tokenizer = Tokenizer(num_words=max_feature, split=' ')
    tokenizer.fit_on_texts(total_comment)
    X = tokenizer.texts_to_sequences(total_comment)
    X = pad_sequences(X)
    X.shape
    
    model = Sequential()
    model.add(Embedding(input_dim = max_feature, 
                        output_dim = 32, 
                        input_length = X.shape[1]))
    
    model.add(LSTM(units = 32,
                   kernel_regularizer=regularizers.l2(0.001),
                   dropout = dropout_rate, 
                   recurrent_dropout = recurrent_dropout_rate
                  ))
              
    options = {"input_dim": X.shape[0]}
    for layer in range(n_hidden_layer):
        model.add(Dense(n_neurons, activation="relu", **options))
        options = {} 
    
    model.add(Dense(n_neurons/2))
    model.add(Dense(2,activation='softmax'))
    
    model.compile(loss = 'categorical_crossentropy', 
                      optimizer = optimizer_mode, 
                      metrics=['accuracy']
                     )
    return model

batch_size = 32
```

```python
model_a = build_model(total_comment = df_all['preprocessed'].values, n_hidden_layer=1, 
                      n_neurons=16, optimizer_mode='Adam', dropout_rate=0.5, recurrent_dropout_rate=0.5)
model_a.summary()
```

### Test and Report
- Make summary report

```python
# Make a model summary report
def model_report_ng(X_test,y_test,GridSearchCV_List,model_func):

    model_list = []
    accuracy_list = []
    balanced_acc_list = []
    auc_list = []
    cm_list = []
    recall_list = []
    precision_list = []
#     precision_pos_list = []
    precision_neg_list = []
    ap_list = []
    f1_list = []
    best_params_list = []

    for i,j in enumerate(GridSearchCV_List):
        locals()[j] = model_func[i]
        
        if isinstance(model_func[i], 
                      (sklearn.model_selection._search.RandomizedSearchCV,
                       sklearn.model_selection.GridSearchCV)):            
            y_pred = eval(j+'.best_estimator_.predict(X_test)')
            P_true = eval(j+'.predict_proba(X_test)[:, 1]')
        else:
            y_pred = eval(j+'.predict_classes(X_test)')
            P_true = eval(j+'.predict_proba(X_test)[:, 1]')
             
        model_name = j
        TP, FN, FP, TN = confusion_matrix(y_test, y_pred, labels=[1, 0]).ravel()
        
        Accuracy = round(accuracy_score(y_test, y_pred), 2)
        bal_acc = round(balanced_accuracy_score(y_test, y_pred),2)
        auc = round(roc_auc_score(y_test, P_true), 2)
        cm = dict(TP=TP, FP=FP, FN=FN, TN=TN)
        Recall = round(recall_score(y_test, y_pred, average='binary'), 2)
        Precision = round(precision_score(y_test, y_pred, average='binary'), 2)
#         Precision_pos = round(precision_score(y_test, y_pred, pos_label = 1, average='binary'), 2)
        Precision_neg = round(precision_score(y_test, y_pred, pos_label = 0, average='binary'), 2)
        ap = round(average_precision_score(y_test, P_true, pos_label=1), 2)
        F1_score = round(f1_score(y_test, y_pred,average='binary'), 2)
#         best_params = eval(j+'.best_params_')
        
        model_list.append(model_name)
        accuracy_list.append(Accuracy)
        balanced_acc_list.append(bal_acc)
        auc_list.append(auc)
        cm_list.append(cm)
        recall_list.append(Recall)
        precision_list.append(Precision)
#         precision_pos_list.append(Precision_pos)
        precision_neg_list.append(Precision_neg)
        ap_list.append(ap)
        f1_list.append(F1_score)
#         best_params_list.append(best_params)
        

    report = dict(Model = model_list,
                  Accuracy = accuracy_list,
                  Balanced_Accuracy = balanced_acc_list,
                  AUC = auc_list,
                  Confusion_Matrix = cm_list,
                  Recall = recall_list,
                  Precision = precision_list,
#                   Precision_pos = precision_pos_list,
                  Precision_neg = precision_neg_list,
                  Average_Precision = ap_list,
                  F1_score = f1_list,
#                   Best_Parameters = best_params_list 
                 )
    
    df_report = pd.DataFrame.from_dict(report)
    pd.set_option('display.max_colwidth', -1)
    
    return df_report
```

## The Error Analysis Contains 6 Type Categories.

1. **Mistranslation**
    - Due to we use Google Translate API to translate comment from Thai into English. In some case, this process cannot perfectly translate and preserve the original meaningful that leaves the model with a misleading understanding. For example; row index 13 'เลือกไม่ผิดที่ไว้ใจและฝากลูกน้อยไว้กับการคลอดที่โรงพยาบาลนี้' is translated into 'Choose not to trust them and leave the baby with hospital', row index 18 'อยากถามว่าฝากท้องเสียค่าใช่จ่ายเท่ารัยคะ' is translated into 'Asked to leave Basra cost me anything.'.
    - Incorrect meaning: row index 10 'หาหมอคลีนิคนอกเวลา ไปตามนัดหมาย วัดความดัน ได้บัตรคิว ไปนั่งรอหน้าห้องหมอ' is translated into 'According to the doctor, clinic appointments outside pressure had to queue waiting rooms, doctors.' In fact, the word 'pressure' in this case is mean 'blood pressure' which is different meaning from 'pressure' due to Thai culture usually say just 'measure pressure' instead of 'measure blood pressure' in full sentence. Row index 4 'ผิดกับที่คิดไว้คับเจ้าหน้าที่ดีมาก...' is translated into 'I think that is wrong with the staff great...'. In fact, this sentence is a positive meaning, but it is translated into negative meaning, which leads to a negative class. The word 'คับ' is translated into 'tight' in spite of no meaning indeed.  
    - Thai abbreviations: We don't see this issue much in this study.
    - No full stop or period in Thai language: The full stop is used for several purposes, the most frequent of which is to mark the end of a declaratory sentence. However, for Thai language, there are no full stops at the end of sentences.  This makes translation more difficult for correctly meaning.
        
- **Sarcastic word**
    - Sarcastic or ridicules, usually by saying the opposite of what is actually meant. For example; row index 7 'บริการเยี่ยม' has direct meaning is great service, but it could be sarcastic because of scoring 2. The model is not intelligent enough to distinguish this type.
- **Incorrect score**
    - Row index 21 the comment is negative meaning. However, the customer gives five scores for this comment.
- **Irrelevant content**
    - Customers come to inquire about the use of various services which is not related to hospital services satisfaction or complaints, which can affect the model. For example; row index 18 is only asking cost service.
- **Spelling errors**
    - For example; row index 1 'ตรวจ' > 'ตรวต', 'หมอ' > 'หมอก'
- **Having both complaint and admire in the same comment**
    - This might make model be confused by interpreting positive word as negative or vice versa.

# Switching Data Source
<p style="text-indent: 2.5em;">
Due to using Google translation can perform many methods such as Google API or google sheet. Although using google sheet can perform smarter conversion than Google API, each comment would have the different word and more unique compared to another comment which has similar content. For example; <i>'บริการค่อยข้างดีค่ะ แต่รอคิวนานนิดหน่อย'</i>, Google API will translate into <i>'Service is good. But wait for a long time'</i> but Google sheet will translate into <i>'Service pretty good But wait a little longer'</i>. This can be an issue for performing word frequency or word cloud. # Word Cloud Visualization
[Top](#ch0)

```python
def word_cloud(df, name_comment='comment_eng', class_show = None, name_class = 'sentiment', Fignum = Fignum ):
    
    # Check condition for word cloud visualization
    if class_show == None:
        title = 'All Sentiment(+/-)'
        array = [1,0]
    elif class_show == 1:
        title = 'Positive Sentiment(+)'
        
        array = [class_show]
    elif class_show == 0:
        title = 'Negative Sentiment(-)'
        array = [class_show]
        
    if len(df['hospital'].unique()) != 1:
        hosp = 'Three Hospitals'
        hospital = '\n'
    else:
        hosp = df['hospital'].unique()[0]
        hospital = ' Hospital\n'
        
    df_show = df.loc[df[name_class].isin(array)]
    df_show = df_show.reset_index(drop=True)
    
    # Eliminate some word that less meaningful
    elim_wd = ['hospital', 'medical team', 'doctor', 'nurse', 'doctor nurse', 'service', 
               'ramathibodi hospital', 'rama hospital', 'chula hospital', 'chulalongkorn hospital', 'siriraj hospital', 
               'state hospital']
    
    # Make collection of noun phrase
    noun_phase_list = []    
    for i in range(len(df_show[name_comment])):
        np = TextBlob(df_show[name_comment][i].lower(), np_extractor=extractor).noun_phrases
        noun_phase_list.extend(np)
        
    # Filtering only noun phrase frequency more than 1
    noun_phase_list = [x for x in noun_phase_list if x not in elim_wd]
    print("Number of unique noun phrase:", len(noun_phase_list))
    print("The most common noun phrase (Frequency > 1):")
    noun_phase_list

    dict_np = Counter(noun_phase_list)
    dict_np_r = Counter(el for el in dict_np.elements() if dict_np[el] > 1)
    sorted_dict_np = sorted(dict_np_r.items(), key = lambda kv: kv[1], reverse=True)
    display(sorted_dict_np[:10])
    
    # Display visualization
    plt.figure(figsize=(12,12))
    wc = WordCloud(background_color="white",width=1000,height=600, max_words=50,relative_scaling=0.5,normalize_plurals=False).generate_from_frequencies(dict_np_r)
    plt.imshow(wc)
    plt.title('Figure '+ str((next(Fignum))) + ': Word Cloud of ' + title + ' in ' + hosp + hospital, fontsize=20)
    plt.axis("off")
    plt.show()
```

### Conclusions
<p style="text-indent: 2.5em;">
    This study is a sentiment analysis that focuses on how to manage complaint or dissatisfaction from the patient. The model evaluation should be based on negative class rather than positive class. The possible metrics to evaluate this model should mainly base on precision on negative class. As the result from the model evaluation, the best model is the model_a which performs the highest precision on negative class (72%). For text understanding, the most ten common noun phrase from the negative sentiment of three hospitals are 'long time', 'good service', 'first time', '3rd floor', '1st floor', 'queue card', 'emergency department', 'bad service', 'emergency room', 'bad feeling'.
    
### Suggestions
<p style="text-indent: 2.5em;">
    <b>For Ramathibodi hospital</b>, the top issue that customer complaint is about "long time" which means many customers have to wait for a long time for any of service. That indicates that this hospital has to improve workflow to be better performance. They also complaint a lot with the emergency room or emergency department — most of this complain about service or healthcare practitioner manner, which is terrible service. The parlance of staff in any department is very bad as well.
<p style="text-indent: 2.5em;">
    <b>For Chulalongkorn hospital</b>, the biggest issue still is the same as Ramathibodi hospital is long wait time especially, at third floor has a problem with waiting time.
<p style="text-indent: 2.5em;">
    <b>For Siriraj hospital</b> same as previous the issue is a wait time.

>***SUMMARY :*** The observed complaints of negative sentiment from customers in three hospitals were found that the enormous issue is the time for waiting and staff' service mind. When drilling into all the details will be seen that this happens in every part of the service, whether it is front desk auscultation, outpatient examination service, dispensing room. Especially, Ramathibodi Hospital, there were complaints a lot about the service at the department of emergency issue. This is the biggest issue which the hospital should check and figure out the cause of the problem and improve it. The analysis of the problem, we anticipate that the leading cause of wait time due to the number of patients with large volume compared to the number of medical practitioners and staff which make the hospital can not provide promptly. With limited resources at the hospital, it is difficult to improve the service even better. 

However, we can improve this issue by our providing solution as below;
1. Improving the workflow for making better service performance such as managing queues, having an application that tells the status of service which can let customer does not keep waiting for servicing.
2. Hire more medical practitioners.
3. To educate people for healthcare knowledge which make them can take care of themselves in order to prevent the disease. This can help to decrease the patient density in hospital.
4. Decentralize healthcare service to public health center or primary/secondary care and promote people with less severity disease to go there instead of the hospital, which is tertiary care.

>There have many suggestion which is very useful for improve hospital service which you can find in this [link](https://www.qminder.com/reduce-hospital-wait-times/) that including 'Invest in EHR and Gather Data in Advance', 'wasted time on medical paperwork', 'the support staff in the waiting area should attend to customers', etc.

<p style="text-indent: 2.5em;">

# Make data from sklearn to DataFrame

```python
# Import Data
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

# save load_iris() sklearn dataset to iris
# if you'd like to check dataset type use: type(load_iris())
# if you'd like to view list of attributes use: dir(load_iris())
iris = load_iris()

# np.c_ is the numpy concatenate function
# which is used to concat iris['data'] and iris['target'] arrays 
# for pandas column argument: concat iris['feature_names'] list
# and string list (in this case one string); you can make this anything you'd like..  
# the original dataset would probably call this ['Species']
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
df.head()
```

```python
# Change column name for easy to understand
df.rename(columns={'target': 'species'}, inplace=True)
df.head(10)
```

```python
iris
print(iris.target_names)
```

```python
df['species'] = df['species'].map({ 0 : 'setosa', 1 : 'versicolor', 2 : 'virginica'})
df.head(10)
```

# How to plot boxplot

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
# Import necessary modules
import matplotlib.pyplot as plt
import seaborn as sns
```

- Begin with one boxplot

```python
plt.boxplot(df['sepal length (cm)'])

# Label the axes
plt.xlabel('sepal length (cm)')
plt.ylabel('cm')
plt.title('sepal length (cm)')

plt.show()
```

```python
# Set size
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.boxplot(df['sepal length (cm)'])

# Label the axes
plt.xlabel('sepal length (cm)')
plt.ylabel('cm')
plt.title('sepal length (cm)')
# fig.savefig('test2png.png', dpi=100)
plt.show()
```

```python
plt.boxplot('sepal length (cm)', data=df)

# Label the axes
plt.xlabel('sepal length (cm)')
plt.ylabel('cm')
plt.title('sepal length (cm)')

plt.show()
```

- If we want to plot more than one in same output

```python
# Create a figure with 2x2 subplot layout and make the top left subplot active
plt.subplot(2,2,1) # two first argument is shape of subplot grid (2,2) = 2 row, 2 columns ,Third argument is position
plt.boxplot(df['sepal length (cm)'])

# Label the axes
plt.xlabel('sepal length (cm)')
plt.ylabel('cm')
plt.title('sepal length (cm)')


# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2,2,2) 
plt.boxplot(df['sepal width (cm)'])

# Label the axes
plt.xlabel('sepal width (cm)')
plt.ylabel('cm')
plt.title('sepal width (cm)')


# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2,2,3)
plt.boxplot(df['petal length (cm)'])

# Label the axes
plt.xlabel('petal length (cm)')
plt.ylabel('cm')
plt.title('petal length (cm)')


# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2,2,4)
plt.boxplot(df['petal width (cm)'])

# Label the axes
plt.xlabel('petal width (cm)')
plt.ylabel('cm')
plt.title('petal width (cm)')

# Improve the spacing between subplots and display them
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Set the style to 'ggplot'
plt.style.use('ggplot')

# Create a figure with 1x2 subplot layout and make the top left subplot active
plt.subplot(2,2,1) # two first argument is shape of subplot grid (2,2) = 2 row, 2 columns ,Third argument is position
plt.boxplot(df['sepal length (cm)'])

# Label the axes
plt.xlabel('sepal length (cm)')
plt.ylabel('cm')
plt.title('sepal length (cm)')


# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2,2,2) 
plt.boxplot(df['sepal width (cm)'])

# Label the axes
plt.xlabel('sepal width (cm)')
plt.ylabel('cm')
plt.title('sepal width (cm)')


# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2,2,3)
plt.boxplot(df['petal length (cm)'])

# Label the axes
plt.xlabel('petal length (cm)')
plt.ylabel('cm')
plt.title('petal length (cm)')


# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2,2,4)
plt.boxplot(df['petal width (cm)'])

# Label the axes
plt.xlabel('petal width (cm)')
plt.ylabel('cm')
plt.title('petal width (cm)')

# Improve the spacing between subplots and display them
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Set the style to 'Seaborn'
sns.set()

# Create a figure with 2x2 subplot layout and make the top left subplot active
plt.subplot(2,2,1)
# Create box plot with Seaborn's default settings
sns.boxplot(y='sepal length (cm)', data=df)

# Label the axes
plt.xlabel('species')
plt.ylabel('sepal length (cm)')
plt.title('sepal length (cm)')

# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2,2,2)
sns.boxplot(y='sepal width (cm)', data=df)
plt.title('sepal width (cm)')

# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2,2,3)
sns.boxplot(y='petal length (cm)', data=df)
plt.title('petal length (cm)')

# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2,2,4)
sns.boxplot(y='petal width (cm)', data=df)
plt.title('petal width (cm)')

# Improve the spacing between subplots and display them
plt.tight_layout()

# Show the plot
plt.show()
```

- Draw a box plot to show distributions with respect to categories.

```python
# Create a figure with 2x2 subplot layout and make the top left subplot active
plt.subplot(2,2,1)
# Create box plot with Seaborn's default settings
sns.boxplot(x='species', y='sepal length (cm)', data=df)

# Label the axes
plt.xlabel('species')
plt.ylabel('sepal length (cm)')
plt.title('sepal length (cm)')

# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2,2,2)
sns.boxplot(x='species', y='sepal width (cm)', data=df)
plt.title('sepal width (cm)')

# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2,2,3)
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.title('petal length (cm)')

# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2,2,4)
sns.boxplot(x='species', y='petal width (cm)', data=df)
plt.title('petal width (cm)')

# Improve the spacing between subplots and display them
plt.tight_layout()

# Show the plot
plt.show()
```

# Multiple box plots on one Axes
- If you want to plot all result in same graph for comparing value

```python
data = df.copy()
DATA = [data['sepal length (cm)'], data['sepal width (cm)'], data['petal length (cm)'], data['petal width (cm)']]

# Multiple box plots on one Axes
fig, ax = plt.subplots()
ax.boxplot(DATA,labels=('sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'))

plt.title('Iris Boxplot',fontsize=20)
plt.show()
```

- Improve the spacing between subplots and display them

plt.tight_layout()

```python
data = df.copy()
DATA = [data['sepal length (cm)'], data['sepal width (cm)'], data['petal length (cm)'], data['petal width (cm)']]

# Multiple box plots on one Axes
fig, ax = plt.subplots()
ax.boxplot(DATA,labels=('sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'))

plt.title('Iris Boxplot',fontsize=20)
# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```

```python
# Changes to the Classic style
plt.style.use('classic')

data = df.iloc
DATA = [data[:,0], data[:,1], data[:,2], data[:,3]]
# Multiple box plots on one Axes
fig, ax = plt.subplots()
ax.boxplot(DATA,labels=('Feature[0]','Feature[1]','Feature[2]','Feature[3]'))

plt.title('Original Data',fontsize=20)
plt.show()
```

```python
# Changes to the default style
plt.style.use('default')

data = df.iloc
DATA = [data[:,0], data[:,1], data[:,2], data[:,3]]
# Multiple box plots on one Axes
fig, ax = plt.subplots()
ax.boxplot(DATA,labels=('Feature[0]','Feature[1]','Feature[2]','Feature[3]'))

plt.title('Original Data',fontsize=20)
plt.show()
```
